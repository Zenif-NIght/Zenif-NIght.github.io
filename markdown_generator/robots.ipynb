```python
import pandas as pd
import os

# Define the HTML escape table for special characters
html_escape_table = {
    "&": "&amp;",
    '"': "&quot;",
    "'": "&apos;"
}


```

```python

# Function to escape special characters in text
def html_escape(text):
    return "".join(html_escape_table.get(c, c) for c in text)

# Load the robot projects TSV file
tsv_file_path = "robot_projects.tsv"
robot_projects = pd.read_csv(tsv_file_path, sep="\t")
robot_projects
```

```python

# Generate markdown files
for index, project_row in robot_projects.iterrows():
    
    md_filename = f"{project_row['date']}-{project_row['url_slug']}.md"
    html_filename = f"{project_row['date']}-{project_row['url_slug']}"
    year = project_row['date'][:4]
    
    # Start YAML front matter
    md = "---\n"
    # md = "---\ntitle: \""   + item.title + '"\n'
    md += f"layout: portfolio\n"
    md += f"title: {project_row['title']}\n"
    md += f"date: {project_row['date']}\n"
    md += f"year: {year}\n"
    md += f"venue: {project_row['venue']}\n"
    md += f"location: {project_row['location']}\n"

    for col in project_row.index:
        cur_value = project_row[col]
        if pd.isna(cur_value):
            continue
        md += f"{col}: {html_escape(str(project_row[col]))}\n"
    md += f"collection: portfolio\n"

    # Close YAML front matter
    md += "---\n"
    
    # Add the description section
    md += f"\n{html_escape(project_row['excerpt'])}\n"
        
    
    if len(str(project_row['video_url'])) > 5:
        md += f"\n<iframe width='560' height='315' src='{project_row['video_url']}' frameborder='0' allowfullscreen></iframe>\n"
    

    # Save markdown file
    output_dir = "../_portfolio/"
    os.makedirs(output_dir, exist_ok=True)
    md_filepath = os.path.join(output_dir, md_filename)
    with open(md_filepath, 'w') as f:
        f.write(md)

md_filepath
```
